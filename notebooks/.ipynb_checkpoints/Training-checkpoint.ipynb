# Stage 6: Model Training and Saving

```python
# defining a function to train a model for each city in the data set
def train_city_model(df, city_name):
    print(f"\nTraining models for {city_name}...")

    # Filtering city data
    city_df[df["City"] == city_name].copy()

    # Creating an if statement to make sure that the number of datapoint to be train for any city is below 1000
    if len(city_df) < 1000:
        print("Not enough data, skipping...")
        return
        
    # Sorting the dataset for that city by time
    city_df = city_df.sort_values("Date")

    # Feature engineering
    city_df["Month"] = city_df["Date"].dt.month
    city_df["Dayofweek"] = city_df["Date"].dt.dayofweek

    for col in ["PM2.5", "PM10"]:
        city_df[f"{col}_lag1"] = city_df[col].shift(1)

    city_df = city_df.dropna()

    # Features & targers
    features = [
        "PM2.5", "PM10", "NO2", "SO2", "CO", "O3", 
        "Month", "Dayofweek", "PM2.5_lag1", "PM10_lag1"
    ]
    # Mapping the output of the AQI_Bucket
    aqi_mapping = {
        "Good": 0,
        "Satisfactory": 1,
        "Moderate": 2,
        "Poor": 3,
        "Very Poor": 4,
        "Severe": 5
    }

    inverse_mapping = {v: k for k, v in aqi_mapping.items()}
    
    X = city_df[features]
    y_reg = city_df["AQI"]
    y_clf = city_df["AQI_Bucket"].map(aqi_mapping)

    # Time-based split
    split_idx = int(len(city_df) * 0.8)

    X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
    y_reg_train, y_reg_test = y_reg.iloc[:split_idx], y_reg.iloc[split_idx:]
    y_clf_train, y_clf_test = y_clf.iloc[:split_idx], y_clf.iloc[split_idx:]

    # Train regression model for the AQI
    rf_reg = RandomForestRegressor(
        n_estimators=500,
        max_depth=25,
        min_samples_leaf=2,
        min_samples_split=5,
        random_state=42,
        n_jobs=-1
    )

    rf_reg.fit(X_train, y_reg_train)
    
    y_pred_reg = rf_reg.predict(X_test)
    mae = mean_absolute_error(y_reg_test, y_pred_reg)
    r2 = r2_score(y_reg_test, y_pred_reg)

    # Train classification model for the AQI_Bucket
    y_clf_train_enc = y_clf_train.values
    y_clf_test_enc = y_clf_test.values
    
    rf_clf = RandomForestClassifier(
        n_estimators=400,
        max_depth=20,
        min_samples_leaf=2,
        min_samples_split=5,
        class_weight="balanced",
        random_state=42,
        n_jobs=-1
    )
    
    rf_clf.fit(X_train, y_clf_train_enc)
    y_pred_clf = rf_clf.predict(X_test)

    labels = np.unique(np.concatenate([y_clf_test_enc, y_pred_clf])) 


    print("Regression  MAE:", round(mae, 2), "| R:", round(r2, 3))
    print("Classification Report:")
    print(classification_report(y_clf_test_enc, 
                                y_pred_clf, 
                                labels=labels,
                                target_names=[inverse_mapping[i] for i in sorted(inverse_mapping)]
                               ))

    # 8 Save models
    joblib.dump(rf_reg, f"../models/{city_name}_rf_reg.pkl")
    joblib.dump(rf_clf, f"../models/{city_name}_rf_clf.pkl")

    print(f" Models saved for {city_name}")
```

```python

# Deploying the training model to check and train model for each 
# city in the data set and skip the once that are not up to 1000 datapoints

cities = df["City"].unique()

for city in cities:
    train_city_model(df, city)
```
