```python
import pandas as pd
import joblib 
import os 
from sklearn.ensemble import RandomForestRegressor
```

```python
DATA = "../data/station_day_with_station_info.csv"
MODEL_DIR = "models"
os.makedirs(MODEL_DIR, exist_ok=True)
```

```python
pollutants = ["PM2.5", "PM10", "NO2", "SO2", "CO", "O3"]
```

```python
df = pd.read_csv(DATA, parse_dates=["Date"])
```

```python
df["Month"] = df["Date"].dt.month
df["Dayofweek"] = df["Date"].dt.dayofweek
```

```python
pollutants
```

```python
for city in df["City"].unique():
    city_df = df[df["City"] == city].sort_values("Date")

    for p in pollutants:
        city_df[f"{p}_lag1"] = city_df[p].shift(1)

        train_df = city_df.dropna()

        X = train_df[[f"{p}_lag1", "Month", "Dayofweek"]]
        y = train_df[p]

        model = RandomForestRegressor(
            n_estimators=300,
            max_depth=15,
            random_state=42
        )
        model.fit(X, y)

        joblib.dump(
            model,
            f"{MODEL_DIR}/{city}_{p}_rf.pkl"
        )

print("✅ Pollutant models trained")
```
